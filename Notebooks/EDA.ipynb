***Performing Exploratory Data Analysis on E-Commerce Data***

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(fr"../synthetic_ecommerce_data.csv")  # Reading dataset from CSV file
```

```python
sns.countplot(data=df, x='purchased')
plt.title("Purchase Distribution")
plt.show()
```

```python
sns.boxplot(data=df, x='gender', y='session_time')
plt.title("Session Time by Gender")
plt.show()
```

```python
df['gender'].value_counts().plot.pie(autopct='%1.1f%%', title='Gender Distribution')
plt.show()
```

```python
df['age'].plot.hist(bins=10, title='Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(13, 27))
df['country'].value_counts().plot.barh(title='Customers by Country')
plt.yticks(fontsize=7)  # Try values like 8, 10, 12
plt.show()
```

```python
df['purchased'].value_counts().plot.pie(labels=['No', 'Yes'], autopct='%1.1f%%', title='Purchase Rate')
plt.show()
```

```python
#Product category
plt.figure(figsize=(14,8))
df['product_category'].value_counts().head(10).plot.bar(title='Most Clicked Product Categories')
plt.show()
```

All the data above shows that fewer products are purchased compared to those not purchased. This means there's an opportunity to increase customer conversions.
